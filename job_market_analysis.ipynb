```python
df = pd.read_sql('select * from job postings',engine)
```

```python
df = df.drop_duplicates()
```

```python
df = df.fillna("Unknown")   # or use df.dropna() if you want to remove rows
```

```python
df['max_salary'] = df['max_salary'].astype(str).str.replace('$','').str.replace(',','')
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce')
```

```python
df['original_listed_time'] = pd.to_datetime(df['original_listed_time'], errors='coerce')
df['year'] = df['original_listed_time'].dt.year
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df = df.fillna("Unknown")

# Convert salary columns to numeric
for col in ['min_salary', 'med_salary', 'max_salary', 'normalized_salary']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert listed_time to datetime
df['listed_time'] = pd.to_datetime(df['listed_time'], errors='coerce')
df['year'] = df['listed_time'].dt.year
```

```python
# Extract job level from experience column
df['job_level'] = df['formatted_experience_level'].replace({
    'ENTRY_LEVEL': 'Entry',
    'MID_LEVEL': 'Mid',
    'SENIOR_LEVEL': 'Senior',
    'Unknown': 'Other'
})

# Split skills into lists (if comma-separated)
df['skills_list'] = df['skills_desc'].apply(lambda x: x.split(',') if x != "Unknown" else [])

# Normalize work type
df['work_type'] = df['formatted_work_type'].replace({
    'FULL_TIME': 'Full-Time',
    'PART_TIME': 'Part-Time',
    'CONTRACT': 'Contract',
    'Unknown': 'Other'
})
```

```python
df.to_sql("job_postings_cleaned", engine, if_exists="replace", index=False)

print("Cleaned dataset successfully stored in PostgreSQL!")
```
